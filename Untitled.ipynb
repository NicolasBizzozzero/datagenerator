```python
dico = CountryDictionary()
with open("databases_to_merge/countries.txt") as file:
    for line in file:
        symbol, country = line.split("|")
        if country[-1] == "\n":
            country = country[:-1]
            
        dico.add_country(country, symbol)
dico._overwrite(safe=False)
```

```python
countries = df.ix[:, 0].tolist()
cities = df.ix[:, 2].tolist()
```

```python
dico = dict()
for country in set(countries):
    dico[country] = {"cities": []}
```

```python
for country, city in zip(countries, cities):
    dico[country]["cities"].append(city)
```

```python
import json

with open("TRUC.json", 'w') as file:
    json.dump(dico, file, indent=4)
```
